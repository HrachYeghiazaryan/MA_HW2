```python
from Bandit import EpsilonGreedy
from Bandit import ThompsonSampling
from Bandit import comparison
```

# Epsilon Greedy

We first define our class instance and "fit" our algorithm to the bandits with provided rewards

```python
#The passed number into the class doesn't play any role in the algorithm, but is necessary for technical reasons.
epsilon_greedy = EpsilonGreedy(1)
```

```python
epsilon_greedy.experiment()
```

Then we plot the average cumulative reward (a.k.a. win rate convergence) for the algorithm fitted above

```python
epsilon_greedy.plot_performance()
```

To conclude talking about Epsilon Greedy we present some statistics and two more plots

```python
df_epsilon_greedy = epsilon_greedy.report()
```

## Thompson Sampling

We first define our class instance and "fit" our algorithm to the bandits with provided rewards.
In addition, we plot our beliefs of distributions for different moments while running the algorithm.

```python
#The passed number into the class doesn't play any role in the algorithm, but is necessary for technical reasons.
thompson_sampling = ThompsonSampling(1)
```

```python
thompson_sampling.experiment()
```

Then we plot the average cumulative reward (a.k.a. win rate convergence) for the algorithm fitted above

```python
thompson_sampling.plot_performance()
```

To conclude talking about Epsilon Greedy we present some statistics and two more plots

```python
df_thompson_sampling = thompson_sampling.report()
```

## Comparison

Here we do the comparison between two algorithms.

```python
comparison()
```

Visually we see that Epsilon Greedy performed worse than Thompson Sampling, nevertheless, t-test yielded
a high p-value which suggests that we fail to reject the mean equality hypothesis and the means of rewards
from both algorithms are equal.
